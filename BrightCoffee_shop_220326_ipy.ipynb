```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("/Bright Coffee Shop Analysis.csv.xlsx")
```

```python
df
```

```python
df.info()
```

```python
df["total_sales"]=df['unit_price']*df['transaction_qty']
df
```

```python
df["month"]="January"
df["transaction_date"] = pd.to_datetime(df["transaction_date"])
df.loc[df["transaction_date"] > pd.Timestamp("2023-01-31"), "month"] = "February"
df.loc[df["transaction_date"] > pd.Timestamp("2023-02-28"), "month"] = "March"
df.loc[df["transaction_date"] > pd.Timestamp("2023-03-31"), "month"] = "April"
df.loc[df["transaction_date"] > pd.Timestamp("2023-04-30"), "month"] = "May"
df.loc[df["transaction_date"] > pd.Timestamp("2023-05-31"), "month"] = "June"
df
```

```python
import datetime as dt # Import datetime for dt.time objects

df["period"] = "morning"
# Convert the 'transaction_time' column to string first, then to datetime objects with specified format,
# and finally extract only the time component for comparison.
# This handles cases where 'transaction_time' might contain mixed types (strings and datetime.time objects)
# and ensures consistent comparison.
df['transaction_time_only'] = pd.to_datetime(df['transaction_time'].astype(str), format='%H:%M:%S').dt.time

# Define datetime.time objects for comparison thresholds
afternoon_threshold = dt.time(12, 0, 0) # Use 12:00:00 for 'afternoon' start (after 11:59:59)
evening_threshold = dt.time(17, 0, 0)   # Use 17:00:00 for 'evening' start (after 16:59:59)
night_threshold = dt.time(20, 0, 0)     # Use 20:00:00 for 'night' start (after 19:59:59)

# Apply conditions to set the 'period'
df.loc[df['transaction_time_only'] >= afternoon_threshold, 'period'] = 'afternoon'
df.loc[df['transaction_time_only'] >= evening_threshold, 'period'] = 'evening'
df.loc[df['transaction_time_only'] >= night_threshold, 'period'] = 'night'

# Drop the temporary column used for comparison
df = df.drop(columns=['transaction_time_only'])
df
```

```python
df['day_of_week'] = df['transaction_date'].dt.day_name()
df
```

```python
df["weekperiod"] = "weekday"
# No need to convert 'day_of_week' to datetime, as it already contains day names as strings.
# Instead, check if the day name is in the list of weekend days.
weekend_days = ['Saturday', 'Sunday']
df.loc[df["day_of_week"].isin(weekend_days), "weekperiod"] = "weekend"
df
```

```python
df.groupby(["store_location"],as_index=False).agg(store_revenue=("total_sales","sum"))
```

```python
plt.figure(figsize=(10, 5))
# Aggregate the data by store_location to get total sales for each store
store_revenue_df = df.groupby(["store_location"], as_index=False).agg(store_revenue=("total_sales", "sum"))
plt.bar(store_revenue_df["store_location"], store_revenue_df["store_revenue"])
plt.title("Total Store Revenue by Location")
plt.xlabel("Store Location")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_store_locations=(df.groupby("store_location")["total_sales"].sum().sort_values(ascending=False).head(3))
plt.figure(figsize=(3, 5))
plt.bar(top_store_locations.index, top_store_locations.values)
plt.title("Top 3 Store Locations by Revenue")
plt.xlabel("Store Location")
plt.ylabel("total Sales")
plt.show()
```
