```python
import pandas as pd
import numpy as nd
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print(f'User uploaded file "{fn}" with length {len(uploaded[fn])} bytes')
  # Assuming the user wants to read this uploaded file
  df=pd.read_excel(fn)

display(df.head())
```
